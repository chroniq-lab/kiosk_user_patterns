```python
import pandas as pd

df = pd.read_csv('/Users/carolinechizak/kiosk_user_patterns/data/kupdat08_cosmos overall estimates.csv')
print(df.head())
```

```python
# Get summary statistics for obesity and overweight
print("Obesity prevalence stats:")
print(df['Obesity'].describe())
print("\nOverweight prevalence stats:")
print(df['Overweight'].describe())
```

```python
import matplotlib.pyplot as plt

# Obesity prevalence histogram
plt.hist(df['Obesity'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('County Obesity Prevalence')
plt.xlabel('Obesity Prevalence (%)')
plt.ylabel('Number of Counties')
plt.show()

# Overweight prevalence histogram
plt.hist(df['Overweight'], bins=20, color='orange', edgecolor='black', alpha=0.7)
plt.title('County Overweight Prevalence')
plt.xlabel('Overweight Prevalence (%)')
plt.ylabel('Number of Counties')
plt.show()
```
